# Parts of Speech-Tagging for Requests


```python
sen1=sp(u"hi,hey ,hello")
print(sen1.text)
print(sen1[0].pos_)
for word in sen1:
    print(f'{word.text:{12}} {word.pos_:{10}} {word.tag_:{8}} {spacy.explain(word.tag_)}')
num_pos = sen1.count_by(spacy.attrs.POS)
num_pos
print("parts of speech :wordcount")
#word count
for k,v in sorted(num_pos.items()):
    print(f'{k}. {sen1.vocab[k].text:{8}}: {v}')
from spacy import displacy
sen1=sp(u"hi ,hey ,hello")

displacy.render(sen1, style='dep', jupyter=True, options={'distance': 85})

```
